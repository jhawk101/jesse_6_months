```python
nappies = read_file_and_clean("nappies")

nappies["week"] = nappies["week"].clip(upper=25)
```

```python
weekly_naps = nappies.groupby(["Status", "week"]).size().to_frame("size")

sns.lineplot(data=weekly_naps, x="week", y="size", hue="Status")
```

```python
nappies.groupby(["week"]).size().to_frame("size").plot(
    ylim=(0,60),
    title="nappies per week",
    ylabel="nappies changes",
    legend=None
)
```
